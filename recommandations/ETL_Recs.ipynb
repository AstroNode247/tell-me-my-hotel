# Transformation des données
## Nettoyage de l'offering

```python
offering = pd.read_csv('../data/offers.csv')
offering = offering.drop('Unnamed: 0', axis=1)
offering = offering.dropna(axis=1, how='all')
offering = offering.drop_duplicates()
offering = offering.drop(axis=1, columns=['url', 'address.street-address', 'address.postal-code', 'region_id', 'address.region'])
offering.head()
```

```python
categories = list(offering['type'].unique())
create_categories = lambda val: 1 if val == cat else 0

for cat in categories :
        offering[cat] = offering['type'].apply(create_categories)

offering = offering.drop(axis=1, columns='type')
offering.head()
```

```python
def dummy_hotel_class(val):
    try:
        if val == hs:
            return 1
        else:
            return 0
    except AttributeError:
        return 0

hotel_class = offering['hotel_class'].unique()
hotel_class = hotel_class[~np.isnan(hotel_class)]
for hs in np.sort(hotel_class):
    offering['class_' + str(hs)] = offering['hotel_class'].apply(dummy_hotel_class)

offering.head()
```

```python
def create_dummies(df, cols, dummy_na=False):
    for col in cols:
        try:
            df = pd.concat([df.drop(col, axis=1), pd.get_dummies(df[col], dummy_na=dummy_na)], axis=1)
        except:
            continue
            
    return df
```

## Nettoyage de reviews

```python
reviews = pd.read_csv('../data/30p_reviews.csv')
reviews = reviews.drop('Unnamed: 0', axis=1)
reviews = reviews.drop_duplicates()
drop_cols = ['date_stayed', 'via_mobile', 'num_helpful_votes', 'author.num_cities', 'author.num_helpful_votes', 
             'author.num_reviews','author.num_type_reviews', 'author.username', 'author.location'] 
reviews = reviews.drop(drop_cols, axis=1)
reviews = reviews.drop(reviews[reviews['author.id'].isnull()==True].index, axis=0)
reviews['date'] = pd.to_datetime(reviews['date'])
reviews.head()
```

```python
def id_mapper():
    coded_dict = dict()
    cter = 1
    id_encoded = []
    
    for val in reviews['author.id']:
        if val not in coded_dict:
            coded_dict[val] = cter
            cter+=1
        
        id_encoded.append(coded_dict[val])
    return id_encoded

id_encoded = id_mapper()
reviews['author.id'] = id_encoded

reviews.head()
```
